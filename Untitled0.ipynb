### 3.Data Preprocessing and Cleaning
#### - **Handling `TotalCharges`**;
`TotalCharges` might have spaces or empty strings which make it an object type. First, replace these non-numeric entries with NaN (Not a Number) then replace with '0.0' For this dataset, customers with 0 tenure (and thus NaN TotalCharges after coercion) likely haven't been charged yet.
####- **Converting data Type**;
 Convert binary categorical columns (Yes/No, Male/Female) to category type
and also other service-related columns that are categorical.
- Checking for duplicates
- Checking for Missing values NaNs

---

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(df.head())# View the first 5 Rows of the data set

# Convert 'TotalCharges' to numeric and fill NaNs
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Inpute with 0.0 for NaNs
df['TotalCharges'] = df['TotalCharges'].fillna(0.0)

# Convert 'SeniorCitizen' to category type
df['SeniorCitizen'] = df['SeniorCitizen'].astype('category')

# Convert Churn to a more usable format (0 and 1)
df['churn_numeric'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# Convert binary categorical columns to category type
categorical_cols = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
    'PaperlessBilling', 'PaymentMethod', 'churn'
]
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
    else:
        print(f"Column '{col}' not found in the DataFrame.")
print(df.info())
print("First Five Rows after Conversion")
print(df.head())
```

```python
# Check for NaNs again after conversion
print(df.isnull().sum())
print("\n")

# check for duplicate rows
duplicates = df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

Then Calculate and visualize churn rates for each tenure band.

These patterns suggest that customer loyalty tends to strengthen over time, with the early months being the most vulnerable period for churn. ## 5.**Feature Engineering**
The next critical step in preparing the data for modeling is feature engineering. Machine learning algorithms, particularly those in Scikit-learn, require all input features to be numerical. The dataset contains numerous categorical features that hold valuable predictive information but are in a non-numeric (string) format.

To convert this categorical data into a format that the models can understand, One-Hot Encoding was employed.


**The Need for One-Hot Encoding:**

**Numerical Representation**: One-hot encoding transforms each category within a feature into a new binary (0 or 1) column. For example, the Contract feature with categories 'Month-to-month', 'One year', and 'Two year' is converted into new columns like `Contract_One year` and `Contract_Two year`.

**Avoiding False Ordinality**: Unlike other methods like Label Encoding (which assigns a unique integer to each category), one-hot encoding is suitable for nominal categorical features where no intrinsic order exists. Using simple integer labels would imply a false relationship (e.g., that 'Two year' (2) is twice 'One year' (1)), which can mislead linear models and other algorithms that are sensitive to the magnitude of feature values.

**Model Compatibility**: This transformation creates a purely numerical feature set that is compatible with all the planned machine learning models.

The pandas.get_dummies() function was used to implement this transformation. To prevent perfect multicollinearity among the newly created dummy variables, the drop_first=True parameter was set, which drops the first category of each feature, creating k-1 dummy variables for k categories.

```python
# Identify categorical columns for one-hot encoding
# Exclude customerID (identifier) and the original Churn column (we have Churn_numeric)
categorical_cols_to_encode = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
    'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod'
]

print("--- DataFrame columns before one-hot encoding ---")
print(df.columns)
print(f"Shape before one-hot encoding: {df.shape}")
print(df.info()) # To see current dtypes
print("\n")

# --- Apply One-Hot Encoding using pd.get_dummies() ---
df_encoded = pd.get_dummies(df, columns=categorical_cols_to_encode, drop_first=True, dtype=int)

print("--- DataFrame columns after one-hot encoding ---")
print(df_encoded.columns)
print(f"Shape after one-hot encoding: {df_encoded.shape}")
print("\n")

print("--- First 5 rows of the encoded DataFrame ---")
print(df_encoded.head())
print("\n")

print("--- Info of the encoded DataFrame ---")
print(df_encoded.info())
```

## 6.2 **Logistic Regression (Baseline)**

The first model developed was a Logistic Regression classifier. This model was chosen as the baseline for several key reasons:

**Interpretability**: Logistic Regression is a linear model, and its coefficients can provide a clear and understandable indication of the relationship between each feature and the likelihood of churn (holding other features constant). This makes it an excellent tool for initial insights.

**Performance Benchmark**: It provides a robust performance baseline. Any more complex model, such as a Random Forest or XGBoost, should demonstrate a clear improvement over this baseline to justify its added complexity.

**Efficiency**: It is computationally fast to train and evaluate.

To account for the class imbalance in the dataset (fewer churners than non-churners), the `class_weight`='balanced' parameter was utilized. This setting automatically adjusts the model's weights to give more importance to the minority class (churners), which typically helps improve recall for that class. Additionally, the numerical features were standardized using `StandardScaler` to ensure that features with different scales did not disproportionately influence the model's coefficient calculations.

---

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report

# Feature scaling for numerical columns
numerical_features_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

# Initialize the StandardScaler
scaler = StandardScaler()
# Fit the scaler on the training data numerical features and transform them
X_train_scaled_numerical = scaler.fit_transform(X_train[numerical_features_to_scale])
# Transform the test data numerical features using the SAME fitted scaler
X_test_scaled_numerical = scaler.transform(X_test[numerical_features_to_scale])

# Convert scaled arrays back to DataFrames
X_train_scaled_numerical_df = pd.DataFrame(X_train_scaled_numerical, columns=numerical_features_to_scale, index=X_train.index)
X_test_scaled_numerical_df = pd.DataFrame(X_test_scaled_numerical, columns=numerical_features_to_scale, index=X_test.index)

# Combine scaled numerical features with the one-hot encoded categorical features
X_train_categorical = X_train.drop(columns=numerical_features_to_scale)
X_test_categorical = X_test.drop(columns=numerical_features_to_scale) # drop the original numerical features from X_train and X_test

# concatenate the scaled numerical features with categorical features
X_train_processed = pd.concat([X_train_scaled_numerical_df, X_train_categorical], axis=1)
X_test_processed = pd.concat([X_test_scaled_numerical_df, X_test_categorical], axis=1)

print("--- X_train_processed head ---")
print(X_train_processed.head())
print("\n")

#Visualize Scaled Numerical Feature Distributions
print("--- Boxplots of Scaled Numerical Features in Training Data ---")
plt.figure(figsize=(12, 6))
sns.boxplot(data=X_train_scaled_numerical_df, palette="Set2")
plt.title('Distribution of Scaled Numerical Features (Training Set)', fontsize=15)
plt.ylabel('Scaled Value (Standard Deviations from Mean)', fontsize=12)
plt.xlabel('Features', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

### 6.3 Random Forest Classifier
To capture more complex, non-linear relationships in the data that a linear model like Logistic Regression might miss, a Random Forest Classifier was developed next.

Random Forest is a powerful ensemble learning method that operates by constructing a multitude of decision trees during training and outputting the class that is the mode of the classes (classification) of the individual trees.

Key Advantages for this Project:

* Robustness to Overfitting: By averaging the predictions of many individual trees (each trained on a different subsample of the data), Random Forest reduces the risk of overfitting that is common with single decision trees.

* Capturing Non-Linearity and Interactions: It does not assume a linear relationship between features and the target, allowing it to identify complex interaction effects without them needing to be explicitly engineered.

* Feature Importance: The model provides a built-in measure of feature importance, offering a different perspective on which features are most predictive compared to the coefficients from Logistic Regression.

The model was first evaluated with its default parameters to establish a quick performance measure. Subsequently, a comprehensive hyperparameter tuning process using GridSearchCV was conducted to optimize for key metrics and find the best-performing model configuration. The class_weight parameter was also used here to continue addressing the class imbalance.


```python
# Initialize the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced_subsample', oob_score=True, n_jobs=-1)

# Train the model
rf_model.fit(X_train_processed, y_train)
print("Random Forest model trained.")
print(f"Out-of-Bag (OOB) Score: {rf_model.oob_score_:.4f}") # Good estimate of generalization error
print("\n")

#  Prediction
y_pred_rf = rf_model.predict(X_test_processed)
y_pred_proba_rf = rf_model.predict_proba(X_test_processed)[:, 1] # Probabilities for ROC-AUC


# Evaluation
print("--- Random Forest Model Evaluation ---")
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
roc_auc_rf = roc_auc_score(y_test, y_pred_proba_rf)

print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall (Sensitivity): {recall_rf:.4f}")
print(f"F1-Score: {f1_rf:.4f}")
print(f"ROC-AUC Score: {roc_auc_rf:.4f}")
print("\n")

print("--- Classification Report (Random Forest) ---")
print(classification_report(y_test, y_pred_rf))
print("\n")

print("--- Confusion Matrix (Random Forest) ---")
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Not Churn', 'Churn'], yticklabels=['Not Churn', 'Churn'])
plt.title('Confusion Matrix - Random Forest', fontsize=15)
plt.ylabel('Actual', fontsize=12)
plt.xlabel('Predicted', fontsize=12)
plt.show()

#  Feature Importance ---
importances = rf_model.feature_importances_
feature_names = X_train_processed.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

print("\n--- Top 15 Feature Importances (Random Forest) ---")
print(feature_importance_df.head(15))

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(15), palette='viridis')
plt.title('Top 15 Feature Importances - Random Forest', fontsize=15)
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
plt.show()
```

Hyperparameter Tuning for Random Forest

```python
from sklearn.model_selection import train_test_split, GridSearchCV
param_grid_rf = {
    'n_estimators': [100, 200],             # Number of trees
    'max_depth': [10, 20, None],            # Max depth of trees
    'min_samples_split': [2, 5, 10],        # Min samples to split a node
    'min_samples_leaf': [1, 2, 4],          # Min samples at a leaf node
    'class_weight': ['balanced', 'balanced_subsample'], # Options to handle imbalance
    'criterion': ['gini', 'entropy']
}

# Initialize the Random Forest Classifier
rf_for_tuning = RandomForestClassifier(random_state=42, n_jobs=-1)

# Initialize GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf_for_tuning,
                              param_grid=param_grid_rf,
                              cv=3, # Using 3-fold for speed
                              scoring='recall', # Prioritizing recall for the churn class
                              verbose=2,
                              n_jobs=-1) # Use all available cores for grid search

# Fit GridSearchCV to the training data
print("Starting GridSearchCV for Random Forest...")
grid_search_rf.fit(X_train_processed, y_train)
print("GridSearchCV training complete.")
print("\n")

# Get the best parameters and the best score
print("\nBest Parameters found by GridSearchCV:")
print(grid_search_rf.best_params_)
print(f"\nBest Recall score on cross-validation: {grid_search_rf.best_score_:.4f}")

# Use the best estimator found by GridSearchCV for predictions
best_rf_model = grid_search_rf.best_estimator_
print("\n")

# --- Prediction with the best model ---

y_pred_rf_tuned = best_rf_model.predict(X_test_processed)
y_pred_proba_rf_tuned = best_rf_model.predict_proba(X_test_processed)[:, 1]

# --- Evaluation of the Tuned Random Forest Model ---
print("\n--- Tuned Random Forest Model Evaluation ---")
accuracy_rf_tuned = accuracy_score(y_test, y_pred_rf_tuned)
precision_rf_tuned = precision_score(y_test, y_pred_rf_tuned)
recall_rf_tuned = recall_score(y_test, y_pred_rf_tuned)
f1_rf_tuned = f1_score(y_test, y_pred_rf_tuned)
roc_auc_rf_tuned = roc_auc_score(y_test, y_pred_proba_rf_tuned)

print(f"Accuracy: {accuracy_rf_tuned:.4f}")
print(f"Precision: {precision_rf_tuned:.4f}")
print(f"Recall (Sensitivity): {recall_rf_tuned:.4f}") # Key metric we aimed to improve
print(f"F1-Score: {f1_rf_tuned:.4f}")
print(f"ROC-AUC Score: {roc_auc_rf_tuned:.4f}")
print("\n")

print("--- Classification Report (Tuned Random Forest) ---")
print(classification_report(y_test, y_pred_rf_tuned))
print("\n")

print("--- Confusion Matrix (Tuned Random Forest) ---")
conf_matrix_rf_tuned = confusion_matrix(y_test, y_pred_rf_tuned)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf_tuned, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Not Churn', 'Churn'], yticklabels=['Not Churn', 'Churn'])
plt.title('Confusion Matrix - Tuned Random Forest', fontsize=15)
plt.ylabel('Actual', fontsize=12)
plt.xlabel('Predicted', fontsize=12)
plt.show()

# --- Feature Importance from the Tuned Model ---
importances_tuned = best_rf_model.feature_importances_
feature_names_tuned = X_train_processed.columns
feature_importance_df_tuned = pd.DataFrame({'Feature': feature_names_tuned, 'Importance': importances_tuned})
feature_importance_df_tuned = feature_importance_df_tuned.sort_values(by='Importance', ascending=False)

print("\n--- Top 15 Feature Importances (Tuned Random Forest) ---")
print(feature_importance_df_tuned.head(15))

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df_tuned.head(15), palette='viridis')
plt.title('Top 15 Feature Importances - Tuned Random Forest', fontsize=15)
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
plt.show()
```

### 6.4 **XGBoost Classifier**

To explore the potential for further performance gains, the final global model implemented was an XGBoost (Extreme Gradient Boosting) Classifier. XGBoost is a highly efficient and powerful implementation of the gradient boosting algorithm, renowned for its state-of-the-art performance on structured/tabular data like the one used in this project.

**Why XGBoost?**

* **Boosting Methodology:** Unlike Random Forest which builds trees in parallel, XGBoost builds trees sequentially. Each new tree is trained to correct the errors made by the previous ones. This iterative approach often leads to higher accuracy.
* **Built-in Regularization**: XGBoost includes regularization terms (L1 and L2) in its objective function, which helps to control model complexity and prevent overfitting, often making it more robust than a standard gradient boosting implementation.
* **Performance:** It is frequently a top-performing algorithm in machine learning competitions and is widely used in industry for problems like churn prediction.


The approach for XGBoost will be similar to that for Random Forest: an initial model will be trained to establish a baseline, followed by a detailed hyperparameter tuning phase using GridSearchCV to optimize its performance for predicting churn. A key parameter, scale_pos_weight, will be used to explicitly manage the class imbalance by increasing the cost of misclassifying the minority (churn) class.

---

```python
from xgboost import XGBClassifier # Import XGBoost
# Model Initialization and Training (XGBoost) ---
# Calculate scale_pos_weight for handling imbalanced classes
# count(negative instances) / count(positive instances)
neg_count = y_train.value_counts()[0]
pos_count = y_train.value_counts()[1]
scale_pos_weight_val = neg_count / pos_count
print(f"Calculated scale_pos_weight: {scale_pos_weight_val:.2f}")

# Initialize the XGBoost Classifier
xgb_model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc', # Using AUC for evaluation and potential early stopping
    scale_pos_weight=scale_pos_weight_val,
    n_estimators=1000,       # Number of boosting rounds/trees
    learning_rate=0.05,      # Step size shrinkage
    max_depth=5,            # Max depth of a tree
    random_state=42,
    use_label_encoder=False, # Suppress warning
    n_jobs=-1,               # Use all available cores
    early_stopping_rounds=10 # Pass early_stopping_rounds to the constructor
)
# Define the evaluation set for early stopping
# Use the test set as the evaluation set
eval_set_for_fit = [(X_test, y_test)]



print("\nTraining XGBoost model...")
xgb_model.fit(X_train, y_train,
              eval_set=eval_set_for_fit,
              verbose=False)

print("XGBoost model trained.")
print("\n")

# Prediction ---
y_pred_xgb = xgb_model.predict(X_test)
y_pred_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]


# Evaluation ---
print("--- XGBoost Model Evaluation ---")
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
precision_xgb = precision_score(y_test, y_pred_xgb)
recall_xgb = recall_score(y_test, y_pred_xgb)
f1_xgb = f1_score(y_test, y_pred_xgb)
roc_auc_xgb = roc_auc_score(y_test, y_pred_proba_xgb)

print(f"Accuracy: {accuracy_xgb:.4f}")
print(f"Precision: {precision_xgb:.4f}")
print(f"Recall (Sensitivity): {recall_xgb:.4f}")
print(f"F1-Score: {f1_xgb:.4f}")
print(f"ROC-AUC Score: {roc_auc_xgb:.4f}")
print("\n")

print("--- Classification Report (XGBoost) ---")
print(classification_report(y_test, y_pred_xgb))
print("\n")

print("--- Confusion Matrix (XGBoost) ---")
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_xgb, annot=True, fmt='d', cmap='Oranges',
            xticklabels=['Not Churn', 'Churn'], yticklabels=['Not Churn', 'Churn'])
plt.title('Confusion Matrix - XGBoost', fontsize=15)
plt.ylabel('Actual', fontsize=12)
plt.xlabel('Predicted', fontsize=12)
plt.show()

# Feature Importance ---
importances_xgb = xgb_model.feature_importances_
feature_names_xgb = X.columns # Using columns from X as X_train was used for fitting
feature_importance_df_xgb = pd.DataFrame({'Feature': feature_names_xgb, 'Importance': importances_xgb})
feature_importance_df_xgb = feature_importance_df_xgb.sort_values(by='Importance', ascending=False)

print("\n--- Top 15 Feature Importances (XGBoost) ---")
print(feature_importance_df_xgb.head(15))

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df_xgb.head(15), palette='mako')
plt.title('Top 15 Feature Importances - XGBoost', fontsize=15)
plt.xlabel('Importance (Gain)', fontsize=12) # XGBoost importance can be 'weight', 'gain', 'cover'
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
plt.show()
```

## 7. Segmented Model Development
While the global models provide a valuable overall perspective on churn, the Exploratory Data Analysis and the high importance of Contract features in all models strongly suggest that a "one-size-fits-all" approach may not be optimal. Customers on Month-to-Month, One-Year, and Two-Year contracts exhibit fundamentally different behaviors, commitment levels, and baseline churn rates.


To address this and align with the dissertation's primary focus on "Contract and Tenure-Based Churn Dynamics," a segmented modeling strategy was implemented. The rationale for this approach is that by training separate models for each contract type, we can:


1. **Capture Segment-Specific Patterns:** Develop models that are more sensitive to the unique churn drivers within each customer group. For example, the factors that cause a month-to-month customer to churn after 3 months are likely very different from those that cause a two-year contract customer to leave after 60 months.

2. **Improve Predictive Performance:** Potentially achieve higher performance (e.g., better F1-score or recall) for specific segments by allowing the models to learn from more homogeneous data, free from the influence of other segments' patterns.

3. **Generate More Actionable Insights:** Derive feature importance rankings that are directly relevant to each contract type, enabling the business to create highly targeted and effective retention strategies.

## Methodology for Segmented Modeling
The following methodology was applied independently to each of the three contract segments (Month-to-Month, One-Year, and Two-Year):

1. Data Segmentation: The fully preprocessed and one-hot encoded dataset was filtered to create three distinct datasets. The Contract dummy variables, now constant within each subset, were removed from the feature set.
2. Independent Data Splitting and Scaling: For each segment, the data was split into new training and testing sets, using stratification to preserve the churn ratio specific to that segment. A new StandardScaler was then fit only on that segment's training data and applied to both its train and test sets.
3. Segment-Specific Hyperparameter Tuning: The two best-performing global model architectures, Random Forest and XGBoost, were selected for segmented modeling. A new GridSearchCV process was conducted for both algorithms on each of the three segments to find the optimal hyperparameters tailored to the unique data characteristics and class imbalance of that specific group. The F1-score for the positive (churn) class was used as the primary optimization metric.
4. Evaluation and Comparison: Each tuned, segmented model was evaluated on its corresponding test set. The results are presented in the following sections to analyze performance and compare the key churn drivers across the different contract types.

### 7.6 Comparative Analysis of Feature Importances

Beyond predictive performance, the primary value of segmented modeling lies in understanding how the drivers of churn differ across customer groups. By comparing the feature importances from our best-performing segmented models (primarily the tuned XGBoost models), we can uncover these nuanced insights.

7.6.1 Global vs. Segmented Feature Importance

The global models consistently identified `Contract_Two year`, `Contract_One year`, and `tenure` as the most important features. This is logical, as the strongest overall predictor of churn is a customer's level of commitment. However, this high-level view masks the specific reasons why customers within those segments leave.

The segmented models provide a much more granular and actionable view.

7.6.2 Key Churn Drivers by Contract Segment

The feature importance rankings for the tuned XGBoost models revealed a significant shift in churn drivers across the three segments:

** Month-to-Month Segment:**

* Top Drivers: `InternetService_Fiber optic`, `PaymentMethod_Electronic` check, `tenure`, and `TotalCharges`.
* Interpretation: For these customers, the decision to churn is heavily influenced by immediate factors.
  * Service-Specific Risk: Having Fiber Optic is the #1 churn indicator, suggesting potential issues with its price point, service stability, or initial value perception for non-committed users.
  * Financial Sensitivity: The importance of payment method and charge amounts (TotalCharges) highlights a strong sensitivity to cost and transaction ease.
  * Short-Term Loyalty: tenure is a dominant factor, indicating that surviving the first few months is critical for retention in this group.

**One-Year Contract Segment:**

* Top Drivers: `StreamingMovies_Yes`, `InternetService_No internet service`, and `MultipleLines_No phone service`.

* Interpretation: For customers with a one-year commitment, the primary churn drivers shift away from raw tenure towards the specifics of their service package.

  * Value Perception of Add-ons: The high importance of StreamingMovies_Yes suggests that customers with this service are a distinct group, and their churn may be related to a re-evaluation of the package's value, perhaps as their contract nears renewal.

  * Niche Customer Profiles: The high importance of features indicating the absence of primary services (no internet, no phone) suggests that churn within this segment is also driven by these specific, non-standard customer profiles.

**Two-Year Contract Segment:**

* Top Drivers: InternetService_Fiber optic, PaymentMethod_Electronic check, and MonthlyCharges.
* Interpretation: Churn among these most loyal customers is a rare event, and when it happens, it appears to be triggered by specific, high-impact friction points.
  * Premium Service Friction: Similar to the MTM segment, InternetService_Fiber optic re-emerges as the top driver. This implies that even for long-term customers, issues related to this premium service (e.g., cost vs. perceived benefit over time) are a primary reason for leaving.
  * Demographics and Cost: MonthlyCharges and SeniorCitizen status are also key drivers, suggesting that even for loyal customers, price sensitivity can emerge over time, or life changes associated with being a senior citizen can trigger a re-evaluation of services.

**7.6.3 Synthesis and Implications**

This comparative analysis demonstrates a clear narrative of the customer churn lifecycle:

* Early/Flexible Stage (Month-to-Month): Churn is driven by initial experiences, service-specific risks (Fiber Optic), and immediate cost.

* Medium-Term Commitment (One-Year): Churn drivers become more nuanced, focusing on the perceived value of the specific service bundle and add-ons.

* Long-Term Commitment (Two-Year): Churn is rare but often tied to specific friction points with high-cost premium services, payment methods, or changes in financial circumstances (MonthlyCharges).

This detailed, segment-specific understanding is far more actionable than a single global view, allowing a business to tailor its retention strategies precisely to the factors that matter most for each customer group.
